## Architecture 
The idea behind the chosen architectures is to gradually add more or less complexity to the model so as to obtain good results and avoid overfitting. 
We'll start with a simple CNN with two convolutional layers and Two Fully connected layers.
#### Convolutional Layers
The main rules of thumb is:
* The number of channels grow as the dimension of the "image" diminishes.

Since the images of the fashion-MNIST database are relatively small (28*28), we can avoid using strides in the convolutional layers. In order to shrink the image slightly, however, we will use max pool layers.

#### Fully connected layers

The architecture of the multilayer perceptron (MLP) used after the convolutional layers has been determined using a couple of rules of thumb:
* Start with one hidden layer, grow from there. 
* Each layer has half less neurons than the previous one.

Let's describe layer per layer:

The input of the multilayer perceptron is the flattened output from the resulting convolutional layer.

Based on previous trials we've decided to go with two hidden layers. Again, this should have been cross validated given enough time. 

To calculate the number of neurons per hidden layer we apply the second rule of thumb. To do so, we had to know the length of the flattened output layer of the convolutional layers. We created the *get_output_shape* function to calculate the dimension of the resulting outputs per layer (Conv2d or Pool2d).

```python
def get_output_shape(width, height, channels, kernel, stride=1, padding=0):
    out_width = ((width-kernel+2*padding)/stride)+1
    out_height = ((height-kernel+2*padding)/stride)+1
    return int(out_width), int(out_height), channels

n_last_conv = 10 # This is the number of times the last convolution is applied to its inputs.
w, h, c = get_output_shape(28, 28, 8, 3, stride=1) # nn.Conv2d(1, 4, 3, stride=3) grayscale image
print(w, h, c)

w, h, c = get_output_shape(w, h, 8, 2, stride=2) # MaxPool2d(2, 2)
print(w, h, c)

w, h, c = get_output_shape(w, h, 16, 3, stride=1) # conv2d(5, n_last_conv, 5, stride=3)
print(w, h, c)

w, h, c = get_output_shape(w, h, 16, 2, stride=2) # MaxPool2d(2, 2)

print(w, h, c)
print("Convolutional network output:", w*h*c)
```

```python
class Classifier(nn.Module):
    def __init__(self):
        super(Classifier, self).__init__()
        # 1 gray channel, 5 kernels, 5x5 kenel size
        self.conv1 = nn.Conv2d(1, 8, 3, stride=1) 
        self.conv2 = nn.Conv2d(8, 16, 3, stride=1)
        self.fc1 = nn.Linear(5*5*16, 80)
        self.fc2 = nn.Linear(80, 10)
        
        self.pool = nn.MaxPool2d(2, 2)
        
    def forward(self, x):
        # if not self.training: -->convert image to tensor and 
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = x.view(-1, 5*5*16)
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        # Don't need the softmax layer since it is used in the Loss function
        return x
```
